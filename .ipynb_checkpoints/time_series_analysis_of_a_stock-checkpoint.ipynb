```python
import numpy as np
import pandas as pd
import yfinance as yf
import datetime as dt
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.stats.diagnostic as sm_diag
```

Download the time series of the daily evolution of a stock with $dt = 1$min and creating white noise

```python
#download data of stock from yahoo finance
data = yf.download(tickers='^GSPC',start=pd.to_datetime('25 jul 2023'),end=pd.to_datetime('26 jul 2023'),\
                   period='1d', interval='1m')

#create dataframe for stock price, returns and log-returns
stock = pd.DataFrame(data.Close[1:])
stock['returns']=data.Close.pct_change().dropna()
stock['log returns']=np.log(data.Close.pct_change().dropna()+1)

stock.Close.plot()

#create dataframe for the white noise
wn = pd.DataFrame(data = np.transpose(np.random.normal(size=len(stock.index))), 
                  index = stock.index, 
                  columns = ['white noise'])
wn.plot()
```

Calculate ACFs and PACFs for stock, returns and white noise. It tells us that we are in an ARIMA(1,1,0) or ARIMA(0,1,0) situation

```python
#find the ACFs for white noise, stock price and returns
acorr1 = sm.tsa.acf(wn, nlags = 10)
acorr2 = sm.tsa.acf(stock.Close, nlags = 10)
acorr3 = sm.tsa.acf(stock.returns, nlags = 10)

#plot the ACFs using pandas plot function
df = pd.DataFrame(data = np.transpose([acorr1,acorr2,acorr3]), 
                  index = range(11), 
                  columns = ['white noise','close','returns'])
plt1 = df.plot( kind="bar")
plt1.set_xlabel('lags')
plt1.set_ylabel('ACF')

#find the PACFs for white noise, stock price and returns
acorr1 = sm.tsa.pacf(wn, nlags = 10)
acorr2 = sm.tsa.pacf(stock.Close, nlags = 10)
acorr3 = sm.tsa.pacf(stock.returns, nlags = 10)

#plot the ACFs using pandas plot function
df = pd.DataFrame(data = np.transpose([acorr1,acorr2,acorr3]), 
                  index = range(11), 
                  columns = ['white noise','close','returns'])
plt1 = df.plot( kind="bar")
plt1.set_xlabel('lags')
plt1.set_ylabel('PACF')
```

Q-Q plot for the different time series

```python
sm.qqplot(wn['white noise'].values, line ='45')
sm.qqplot((stock['Close'].values-stock['Close'].values.mean())/stock['Close'].values.std(), line ='45')
sm.qqplot(stock['returns'].values/stock['returns'].values.std(), line ='45')
sm.qqplot(stock['log returns'].values/stock['log returns'].values.std(), line ='45')
```

Fitting the stock with AR(1) model and forecasting

```python
arima_model = sm.tsa.ARIMA(stock['Close'].values, order=(1, 1, 0))
fitted_model = arima_model.fit()
print(fitted_model.summary())


# Get the number of periods for the forecast
n_forecast_periods = 100  # Adjust this according to your desired forecast horizon
returns_series = stock['Close'].values

# Forecast returns for the next 'n_forecast_periods' periods
forecast_results = fitted_model.get_forecast(steps=n_forecast_periods)

# Extract the forecasted values and their confidence intervals
forecast = forecast_results.predicted_mean
conf_int = forecast_results.conf_int()

# Plot the original series
plt.figure(figsize=(10, 6))
plt.plot(returns_series, label='Observed', color='blue')

# Plot the in-sample and forecasted values
start_idx = len(returns_series)
end_idx = start_idx + n_forecast_periods
plt.plot(np.arange(start_idx, end_idx), forecast, label='Forecast', color='green')

# Plot the confidence intervals
plt.fill_between(np.arange(start_idx, end_idx), \
    conf_int[:,0], conf_int[:,1], color='lightgray', alpha=0.5, label='95% C.I.')

plt.xlabel('Time')
plt.ylabel('Returns')
plt.title('ARIMA(1,1,0) Model Forecast')
plt.legend()
plt.show()
```

```python
# Get the residuals of the model
residuals = fitted_model.resid[1:]

plt.figure(figsize=(6, 5))
plt.plot(residuals)

sm.qqplot(residuals, line='s')

#plot the ACFs using pandas plot function
acorr1 = sm.tsa.acf(residuals, nlags = 10)
df = pd.DataFrame(data = acorr1, 
                  index = range(11), 
                  columns = ['residuals'])
plt1 = df.plot( kind="bar")
plt1.set_xlabel('lags')
plt1.set_ylabel('ACF')

#plot the ACFs using pandas plot function
acorr1 = sm.tsa.pacf(residuals, nlags = 10)
df = pd.DataFrame(data = acorr1, 
                  index = range(11), 
                  columns = ['residuals'])
plt1 = df.plot( kind="bar")
plt1.set_xlabel('lags')
plt1.set_ylabel('PACF')
```
