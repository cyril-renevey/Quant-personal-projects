# Heston Model:

## under risk-neutral measure:
$$
dS_t= r S_t dt+\sqrt{v_t}S_t dW_{1,t}^\mathbb{Q}
\\
dv_t=\kappa(\theta-v_t)dt+\sigma\sqrt{v_t}dW_{2,t}^\mathbb{Q}
\\
dW_{1,t}^\mathbb{Q}dW_{2,t}^\mathbb{Q}=\rho dt
$$

## Call price using mixed FFT and Quadrature

```python
def heston_charfunc_QFwP(phi, S0, v0, kappa, theta, sigma, rho,lambd, tau, r):
    
    kappaQ=kappa+lambd
    thetaQ=theta*kappa/kappaQ
    
    rspi = rho*sigma*phi*1j
    xi = np.sqrt(sigma**2*(phi**2+1j*phi)+(kappaQ-rspi)**2)
    
    wphi = np.exp(1j*phi*np.log(S0)+1j*phi*(r)*tau+kappaQ*thetaQ*tau*(kappaQ-rspi)/sigma**2)\
                  /(np.cosh(xi*tau/2)+(kappaQ-rspi)/xi*np.sinh(xi*tau/2))
    
    return wphi * np.exp(-((phi**2 + 1j*phi) * v0) / (xi / np.tanh(0.5*xi*tau) + kappaQ - rspi))

def integrand_QFwP(phi,K, alpha, S0, v0, kappa, theta, sigma, rho,lambd, tau, r):
    args = ( S0, v0, kappa, theta, sigma, rho,lambd, tau, r)
    psi = heston_charfunc_QFwP(phi-(alpha+1)*1j, *args)/np.exp(r*tau)/(alpha+1j*phi)/(alpha+1j*phi+1)
    return psi*np.exp(-1j*phi*np.log(K))
    
    
def heston_call_price_QFwP(K,alpha, S0, v0, kappa, theta, sigma, rho,lambd, tau, r):
    args = (K,alpha,S0, v0, kappa, theta, sigma, rho,lambd, tau, r)
    
    integral, err = quad(integrand_QFwP, 0, 1, args=args)
    return np.exp(-alpha*np.log(K))/np.pi*integral
    
```

```python
# Parameters to test model
alpha = np.linspace(0.1,15,200) # damping parameter
S0 = 100. # initial asset price
K = 100. # strike
v0 = 0.1 # initial variance
r = 0.03 # risk free rate
kappa = 1.5768 # rate of mean reversion of variance process
theta = 0.0398 # long-term mean variance
sigma = 0.3 # volatility of volatility
lambd = 0 # risk premium of variance
rho = -0.5711 # correlation between variance and stock process
tau = 1.
# time to maturity

prices=[heston_call_price_QFwP(K,i, S0, v0, kappa, theta, sigma, rho,lambd, tau, r) for i in alpha]
df = pd.DataFrame(data = prices, 
                  index = alpha)
df.plot(legend=0)
```

```python
import numpy as np
from scipy.fft import fft

def heston_characteristic_function(phi, S0, r, tau, kappa, theta, sigma, rho, v0):
    
    # constants
    b = kappa
    a = kappa*theta
    
    # common terms w.r.t phi
    rspi = rho*sigma*phi*1j
    
    # define d parameter given phi and b
    d = np.sqrt( (rho*sigma*phi*1j - b)**2 + (phi*1j+phi**2)*sigma**2 )
    
    # define g parameter given phi, b and d
    g = (b-rspi+d)/(b-rspi-d)
    
    # calculate characteristic function by components
    exp1 = np.exp(r*phi*1j*tau)
    term2 = S0**(phi*1j) * ( (1-g*np.exp(d*tau))/(1-g) )**(-2*a/sigma**2)
    exp2 = np.exp(a*tau*(b-rspi+d)/sigma**2 + v0*(b-rspi+d)*( (1-np.exp(d*tau))/(1-g*np.exp(d*tau)) )/sigma**2)
    return exp1*term2*exp2

def heston_fft_call_price(S0, K, r, T, kappa, theta, sigma, rho, v0, N=4096):
    """
    Calculate the Heston call option price using FFT.
    
    :param S0: Initial stock price.
    :param K: Strike price.
    :param r: Risk-free interest rate.
    :param T: Time to expiration.
    :param kappa: Mean reversion speed.
    :param theta: Long-term variance.
    :param sigma: Volatility of variance.
    :param rho: Correlation between stock and variance processes.
    :param v0: Initial variance.
    :param N: Number of discretization points for FFT.
    :return: Heston call option price.
    """
    # Define the log-strike range for FFT
    x_max = 0.5 * N
    dx = x_max / N
    dk = 2 * np.pi / (N * dx)
    b = x_max + dk / 2
    
    # Generate the range of log-strikes
    log_strikes = np.linspace(-x_max, x_max - dx, N)
    
    # Apply the FFT on the characteristic function
    u = np.arange(0, N) * dk
    cf_values = heston_characteristic_function(u - 1j * 0.5, S0, r, T, kappa, theta, sigma, rho, v0)
    fft_values = fft(cf_values).real
    
    # Obtain the call option prices through FFT
    call_prices = np.exp(-r * T) / np.pi * np.exp(-1j * u * (np.log(S0) - b)) * fft_values
    
    # Interpolate the option prices to get the value at K
    call_price = np.interp(np.log(K), log_strikes, call_prices)
    
    return call_price.real

# Example usage:
S0 = 100. # initial asset price
K = 100. # strike
v0 = 0.1 # initial variance
r = 0.03 # risk free rate
kappa = 1.5768 # rate of mean reversion of variance process
theta = 0.0398 # long-term mean variance
sigma = 0.3 # volatility of volatility
lambd = 0 # risk premium of variance
rho = -0.5711 # correlation between variance and stock process
T = 1. # time to maturity

call_price = heston_fft_call_price(S0, K, r, T, kappa, theta, sigma, rho, v0)
print("Heston call option price:", call_price)
```

## Call price using Quantpy

```python
def heston_charfunc_quantpy(phi, S0, v0, kappa, theta, sigma, rho, lambd, tau, r):
    
    # constants
    b = kappa+lambd
    a = kappa*theta
    
    # common terms w.r.t phi
    rspi = rho*sigma*phi*1j
    
    # define d parameter given phi and b
    d = np.sqrt( (rho*sigma*phi*1j - b)**2 + (phi*1j+phi**2)*sigma**2 )
    
    # define g parameter given phi, b and d
    g = (b-rspi+d)/(b-rspi-d)
    
    # calculate characteristic function by components
    exp1 = np.exp(r*phi*1j*tau)
    term2 = S0**(phi*1j) * ( (1-g*np.exp(d*tau))/(1-g) )**(-2*a/sigma**2)
    exp2 = np.exp(a*tau*(b-rspi+d)/sigma**2 + v0*(b-rspi+d)*( (1-np.exp(d*tau))/(1-g*np.exp(d*tau)) )/sigma**2)
    return exp1*term2*exp2
   
def integrand_quantpy(phi, S0, v0, kappa, theta, sigma, rho, lambd, tau, r):
    args = (S0, v0, kappa, theta, sigma, rho, lambd, tau, r)
    numerator = np.exp(r*tau)*heston_charfunc_quantpy(phi-1j, *args) - K*heston_charfunc_quantpy(phi, *args)
    denominator = 1j*phi*K**(1j*phi)
    return numerator / denominator

def heston_price_quantpy(S0, K, v0, kappa, theta, sigma, rho, lambd, tau, r):
    args = (S0, v0, kappa, theta, sigma, rho, lambd, tau, r)
    
    # Use quadrature for numerical integration with proper bounds
    real_integral, err = quad(integrand_quantpy, 0, 100, args=args)
    
    # Calculate the option price
    option_price = (S0 - K*np.exp(-r*tau))/2 + real_integral/np.pi
    
    return option_price
```

```python
S0 = 100. # initial asset price
K = 100. # strike
v0 = 0.1 # initial variance
r = 0.03 # risk free rate
kappa = 1.5768 # rate of mean reversion of variance process
theta = 0.0398 # long-term mean variance
sigma = 0.3 # volatility of volatility
lambd = 0 # risk premium of variance
rho = -0.5711 # correlation between variance and stock process
tau = 1. # time to maturity

heston_price_quantpy( S0, K, v0, kappa, theta, sigma, rho, lambd, tau, r )
```

## Call price using Monte-Carlo

```python
def heston_model_sims(S0, v0, kappa, theta, sigma, rho, lambd, tau, r, N, M):
    """
    Inputs:
     - S0, v0: initial parameters for asset and variance
     - rho   : correlation between asset returns and variance
     - kappa : rate of mean reversion in variance process
     - theta : long-term mean of variance process
     - sigma : vol of vol / volatility of variance process
     - tau     : time of simulation
     - N     : number of time steps
     - M     : number of scenarios / simulations
    
    Outputs:
    - asset prices over time (numpy array)
    - variance over time (numpy array)
    """
    # initialise other parameters
    
    kappaQ=kappa+lambd
    thetaQ=theta*kappa/kappaQ
    dt = tau/N
    
    # arrays for storing prices and variances
    S = np.full(shape=(N+1,M), fill_value=S0)
    v = np.full(shape=(N+1,M), fill_value=v0)
    
    # sampling correlated brownian motions under risk-neutral measure
    Z1 = np.random.normal(size=(N,M))
    Z2 = rho * Z1 + np.sqrt(1 - rho**2) * np.random.normal(size=(N,M))
    
    for i in range(1,N+1):
        S[i] = S[i-1] * np.exp( (r - 0.5*v[i-1])*dt + np.sqrt(v[i-1] * dt) * Z1[i-1,:] )
        v[i] = np.maximum(v[i-1] + kappaQ*(thetaQ-v[i-1])*dt + sigma*np.sqrt(v[i-1]*dt)*Z2[i-1,:],0)
    
    return S, v

def heston_price_MC(S0, K, v0, kappa, theta, sigma, rho, lambd, tau, r):
    N = 500
    M = 10000 
    S,v = heston_model_sims(S0, v0, kappa, theta, sigma, rho, lambd, tau, r, N, M)
    payoff = np.array([np.exp(-r*tau)*np.maximum(i-K,0) for i in S[-1,:]])
    
    return payoff.mean(),payoff.std()/np.sqrt(M)
```

```python
S0 = 100. # initial asset price
K = 100. # strike
v0 = 0.1 # initial variance
r = 0.03 # risk free rate
kappa = 1.5768 # rate of mean reversion of variance process
theta = 0.0398 # long-term mean variance
sigma = 0.3 # volatility of volatility
lambd = 0 # risk premium of variance
rho = -0.5711 # correlation between variance and stock process
tau = 1. # time to maturity

heston_price_MC( S0, K, v0, kappa, theta, sigma, rho,lambd, tau, r )
```
