```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import pylab
import yfinance as yf
import datetime as dt
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Import data from yahoo finance

```python
data = yf.download(tickers='UBSG.SW',start=pd.to_datetime('7 jul 2023'),end=pd.to_datetime('8 jul 2023'),\
                   period='1d', interval='1m')
close = data.Close
close.plot()
```

Estimate mu and sigma of S(t) using average and std of log-returns, generate one realisation of W_P(t), S_P(t) and plot it

```python
simpleDR=close.pct_change().dropna()

mu=simpleDR.mean()
sig=simpleDR.std()
S0=close[0]
print(S0,mu,sig)

timeRange=np.array(range(len(close)))
Wt=np.random.normal(size=len(close)).cumsum()
St=S0*np.exp((mu-sig**2/2)*timeRange+sig*Wt)

data['St']=St
data['Wt']=Wt
data.St.plot()
```

Check method of averaging on simple returns and averaging on log returns

```python
simpleDR=close.pct_change().dropna()
logDR=np.log(simpleDR+1)

mu=simpleDR.mean()
sig=simpleDR.std()

muT=logDR.mean()
sigT=logDR.std()

S0=close[0]

timeRange=np.array(range(len(close)))
Wt=np.random.normal(size=len(close)).cumsum()
St=S0*np.exp((mu-sig**2/2)*timeRange+sig*Wt)
StT=S0*np.exp(muT*timeRange+sigT*Wt)

data['diff']=StT-St
data['diff'].plot()
```

Plot S_P(t) for different realisation of W_P(t) and calculate average S(T) from model and numerically

```python
timeRange=np.array(range(len(close)))
numberW=range(100)
fullTimeRange=np.array([timeRange.tolist() for i in numberW])

W=np.array([np.random.normal(size=len(close)).cumsum().tolist() for i in numberW])

S=S0*np.exp((mu-(sig**2)/2)*fullTimeRange+sig*W)

df = pd.DataFrame(data = np.transpose(S), 
                  index = timeRange, 
                  columns = [str(x) for x in numberW])




df.plot(legend=0)


print(S0*np.exp(mu*timeRange[-1]),S[:,-1].mean())
```

Calculate call price using E_Q(call payoff), W_Q(t) and monte-carlo sim and comparing to value for BSM formula

```python
K=17.5
r=0.02/365/24/60

timeRange=np.array(range(len(close)))
numberWQ=range(10000)
fullTimeRange=np.array([timeRange.tolist() for i in numberWQ])

WQ=np.array([np.random.normal(size=len(close)).cumsum().tolist() for i in numberWQ])

SQ=S0*np.exp(-sig**2/2*fullTimeRange+sig*WQ)

print(np.array([max([i-K,0]) for i in SQ[:,-1]]).mean()*np.exp(-r*timeRange[-1]))

dp=1/np.sqrt(timeRange[-1])/sig*(np.log(S0/K)+(r+sig**2/2)*timeRange[-1])
dm=dp-sig*np.sqrt(timeRange[-1])
print(norm.cdf(dp)*S0-norm.cdf(dm)*K*np.exp(-r*timeRange[-1]))
print(S0-K*np.exp(-r*timeRange[-1]))
```

Simulate S(t) from the SDE and compare to S(t) from geometric BM with arbitrary mu and sig

```python
Dt=1
mu=0.01
sig=0.01
N=100000
T=1

timeRange=np.array(range(N))/N*T
DW=np.random.normal(loc=0.0, scale=np.sqrt(T/N),size=N-1)
Wt=DW.cumsum()

StGBM=S0*np.exp((mu-sig**2/2)*timeRange[1:]+sig*Wt)
StGBM=np.array([S0]+StGBM.tolist())

StSDE=[S0]
for i in range(1,N):
    StSDE.append(StSDE[i-1]+StSDE[i-1]*(mu*T/N+sig*DW[i-1]))

df = pd.DataFrame(data = np.transpose([StGBM,StSDE]), 
                  index = timeRange, 
                  columns = ['StGBM','StSDE'])

df.plot(legend=0)
```
